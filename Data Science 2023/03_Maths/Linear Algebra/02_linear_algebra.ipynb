## **Linear Algebra**

```python
import numpy as np
```

```python
# Linear combination 

x = np.array([[0,1,1],
              [1,1,0],
              [1,0,1]])
y = np.array([3.65, 1.55, 3.42])

scaler = np.linalg.solve(x,y)
scaler
```

```python
# Linear Dependence and Span
import matplotlib.pyplot as plt

x = np.arange(-10, 10)
y = 2*x + 1

plt.figure()
plt.plot(x, y)
plt.xlim(-2, 10)
plt.ylim(-2, 10)
# draw axes
plt.axvline(x=0, color='#A9A9A9')
plt.axhline(y=0, color='#A9A9A9')
plt.show()
plt.close()


x = np.arange(-10, 10)
y = 2*x + 1
y1 = 6*x - 2

plt.figure()
plt.plot(x, y)
plt.plot(x, y1)
plt.xlim(-2, 10)
plt.ylim(-2, 10)
#draw axes
plt.axvline(x=0, color='#A9A9A9')
plt.axhline(y=0, color='#A9A9A9')
plt.show()
plt.close()
```

```python
# Trace
# the Trace is the sum of diagonal elements in a square matrix

a = np.array([[2, 2, 1],
              [1, 3, 1],
              [1, 2, 2]])
print('a = ')
print(a)
print('Trace : ', a.trace())
print('Trace : ', sum(a.diagonal()))
```

```python
# Rank
# The Rank of a matrix is the dimensions of the vector space spanned (generated) by its column or row
# In other words, it can be defined as  the maximum number of linearly independent column vectors or row vectors

a = np.arange(1, 10)
a.shape = (3, 3)
print('a = ')
print(a)
rank = np.linalg.matrix_rank(a)
print('Rank : ', rank)
```

```python
# Determinant

a = np.array([[2, 2, 1],
              [1, 3, 1],
              [1, 2 , 2]])
print('a = ')
print(a)
det = np.linalg.det(a)
print('Determinant : ', np.round(det))
```

```python
# True Inverse
# If the determinant of square matrix is not 0, then it has a true inverse

a = np.array([[2, 2, 1],
              [1, 3, 1],
              [1, 2 , 2]])
print('a = ')
print(a)
det = np.linalg.det(a)
print('Determinant : ', np.round(det))

inv = np.linalg.inv(a)
print('Inverse of a = ')
print(inv)
```

```python
# If you try to compute the true inverse of singular matrix (a square matrix whose determinant is 0) will give an error

a = np.array([[2, 8],
              [1, 4]])
print('a = ')
print(a)
det = np.linalg.det(a)
print('Determinant : ', np.round(det))

inv = np.linalg.inv(a)
print('Inverse of a = ')
print(inv)
```

```python
# Flatten 
# Flatten is a simple method  to transform  a matrix into one-dimensional numpy array

a = np.arange(1, 10)
a.shape = (3, 3)
print('a = ')
print(a)
print('After Flattening')
print('--------------------')
print(a.flatten())
```

```python
# Eigenvalues and Eigenvectors

a = np.array([[2, 2, 1],
              [1, 3, 1], 
              [1, 2 , 2]])
print('a = ')
print(a)
w, v = np.linalg.eig(a)
print('Eigenvalues : ', w)
print('Eigenvectors : ')
print(v)


# the sum of the eigenvalues (5+1+1 = 7) is equal to the trace (2+3+2 = 7) of the same matrix, The product of Eigenvalues (5x1x1 = 5) is equal to the determinant (5) of the same matrix. 
```
