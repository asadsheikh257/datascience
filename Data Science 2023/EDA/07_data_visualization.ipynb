```python
df_titanic = sns.load_dataset('titanic')
df_titanic.info()
```

```python
sns.set_style('darkgrid')
sns.scatterplot(data=df_titanic, y='fare', x='age', hue='sex', palette='rocket')
plt.ylabel('Fare')
plt.xlabel('Age')
plt.title('Titanic Data Set')
plt.show()
```

### Box plot

```python
sns.set_style('darkgrid')
sns.boxplot(data=df_titanic, y='age', x='sex',
                hue='class', palette='rocket')
plt.ylabel('age')
plt.xlabel('sex')
plt.title('Titanic Data Set')
plt.legend(loc=1)
plt.show()
```

### Boxen Plot

```python
sns.set_style('darkgrid')
sns.boxenplot(data=df_titanic, y='age', x='sex',
            hue='class', palette='rocket')
plt.ylabel('age')
plt.xlabel('gender')
plt.title('Titanic Data Set')
plt.legend(loc=1)
plt.xticks(ticks=[0,1], labels=['Male', 'Female'], rotation=38)
plt.show()
```

### Barplot 

```python
sns.set_style('darkgrid')
sns.barplot(data=df_titanic, y='age', x='sex',
              hue='class', palette='rocket', capsize=0.1)
plt.ylabel('age')
plt.xlabel('gender')
plt.title('Titanic Data Set')
plt.legend(loc=1)
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'], rotation=38)
plt.show()
```

### Violin Plots

```python
sns.set_style('darkgrid')
sns.violinplot(data=df_titanic, y='age', x='sex',
            hue='class', palette='rocket')
plt.ylabel('Age')
plt.xlabel('Gender')
plt.title('Titanic Data Set')
plt.legend(loc=1)
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'], rotation=38)
plt.show()
```

```python
df_iris = px.data.iris()
df_iris.head()
df_iris.info()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_iris['sepal_length'], y=df_iris['sepal_width'], mode='markers', name='sepal'))
fig.add_trace(go.Scatter(x=df_iris['petal_length'], y=df_iris['petal_width'], mode='markers', name='petal'))
fig.update_layout(
    title='Iris dataset on Plotly',
    xaxis_title='Iris Flowers Length',
    yaxis_title='Iris Flowers Width',    
)
fig.show()
```

```python
px.scatter(df_iris, x='sepal_length', y='sepal_width', color='species', facet_col='species', hover_name='sepal_width')
```

```python
df_order = df_iris.sort_values(by=['sepal_length'])
df_order1 = df_iris.sort_values(by=['petal_length'])
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_order.sepal_length, y=df_iris.sepal_width, mode='lines', name='Sepal_Width'))
fig.add_trace(go.Scatter(x=df_order1.petal_length, y=df_iris.petal_width, mode='lines+markers', name='Petal_Width'))
fig.update_layout(
    title='Sepal vs. Petal',
    xaxis = dict(title='Length'),
    yaxis = dict(title='Width'),
    showlegend=True,
    plot_bgcolor='white',
)
fig.show()
```
