# **Searbon Library**

```python
# importing libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Seaborn line plot method

```python
var = [1,2,3,4,5,6]
var_1 = [2,4,6,8,5,4]

xd = pd.DataFrame({'var':var, 'var_1':var_1})

sns.lineplot(x='var', y='var_1', data=xd)

plt.show()
```

```python
df = sns.load_dataset('penguins')

df.head()
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=df, hue='sex',style='sex', palette='rocket', markers=['o','>'], dashes=False)
plt.title('Penguins Bill Length vs Bill Depth')
plt.show()
```

### Histplot

```python
plt.figure(figsize=(10,6))
sns.histplot(df['bill_length_mm'], bins=[30,35,40,45,50,55,60], kde=True)
plt.show()
```

## Barplot

```python
order_1 = ['Biscoe', 'Dream', 'Torgersen']
plt.figure(figsize=(6,4))
sns.barplot(x='island', y='bill_length_mm', data=df, hue='sex', order=order_1)
plt.show()
```

### Scatter Plot

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=df, hue='sex', style='sex', palette='rocket')
plt.show()
```

## HeatMap

```python
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var, annot=True, cmap='rocket')
plt.show()
```

### CounterPlot

```python
var = sns.load_dataset('tips')
var
```

```python
sns.countplot(y='sex', data=var, palette='rocket', hue='smoker')
plt.show()
```

### ViolinPlot

```python
sns.violinplot(x='day', y='total_bill', data=var, hue='time')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=var, linewidth=2)
plt.show()
```

```python
sns.violinplot(x='time', y='total_bill', data=var, order=['Dinner', 'Lunch'], inner='points')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=var, hue='sex')
plt.show()
```

```python
sns.violinplot(x='total_bill', data=var)
plt.show()
```

### PairPlot

```python
var = sns.load_dataset('tips')
var
```

```python
sns.pairplot(var, hue='sex', palette='rocket', markers=['>','*'])
plt.show()
```

### Strip Plot

```python

sns.stripplot(x='day', y='tip', data=var, hue='sex', palette='rocket')
plt.show()
```

### Box plot

```python
sns.set(style='whitegrid')
sns.boxplot(x='day', y='total_bill', data=var, hue='sex', showmeans=True, meanprops={'marker':'+', 'markeredgecolor':'white'}, palette='rocket')
plt.show()
```

### CatPlot

```python
# Options are 'strip', 'swarm', 'box', 'boxen', 'violin', 'bar', 'count', and 'point
sns.catplot(x='day', y='size', data=var, hue='sex', kind='point')
plt.show()
```

### FacetGrid

```python
fg = sns.FacetGrid(var, col='day', hue='sex')
fg.map(plt.scatter, 'total_bill', 'tip').add_legend()
plt.show()
```
