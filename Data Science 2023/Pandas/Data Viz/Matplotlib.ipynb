```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### BarPlot

```python
x = ['Python', 'C', 'C++', 'Java']
y = [85,45,65,76]
plt.xlabel('Programming Languages')
plt.ylabel('No:')
plt.title('WSCube')
plt.bar(x,y)
plt.show()
```

```python
x = ['Python', 'C', 'C++', 'Java']
y = [85, 45, 65, 76]
z = [20, 30, 40, 50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel('Programming Languages')
plt.ylabel('No:')
plt.title('WSCube')

plt.bar(p, y, width, label='Popularity')
plt.bar(p1, z, width, label='Popularity_1')

plt.xticks(p+width/2, x, rotation=15)
plt.legend()
plt.show()
```

### PiePlot

```python
x = [10,20,30,40]
y=['c', 'c++', 'java', 'python']
ex = [0.2, 0.0, 0.0, 0.0]

plt.pie(x, labels=y, autopct='%0.1f%%', explode=ex, shadow=True, counterclock=False)
plt.title('wscube tech')
plt.legend(loc=2)
plt.show()
```

```python
x = [10, 20, 30, 40]
y = ['c', 'c++', 'java', 'python']


plt.pie(x, labels=y)
cr = plt.Circle(xy=(0,0), radius=0.6, facecolor='w')
plt.gca().add_artist(cr)
plt.title('wscube tech')
plt.legend(loc=2)
plt.show()
```

### Stem Plot

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt='-', markerfmt='>', basefmt='g', label='Python', orientation='horizontal')
plt.legend()
plt.show()
```

### Boxplot vs Whisker Plot

```python
x = [10,20,30,40,50,60,70,120]


plt.boxplot(x, labels=['Python'], patch_artist=True, showmeans=False), 
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]
y = [30, 40, 50, 60, 70, 80, 90]

data = [x, y]

plt.boxplot(data, labels=['C++', 'Python'], patch_artist=True),
plt.show()
```

### Stack Plot

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

label = ['Area 1', 'Area 2', 'Area 3']

plt.stackplot(x, area1, area2, area3, labels=label)
plt.title('Python')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc=2)
plt.show()
```

### Step Plot

```python
x = [1,2,3,4,5,6]
y = [2,5,4,1,6,3]

plt.step(x,y, marker='o', label='Python')
plt.title('Python')
plt.xlabel('x-axis')
plt.ylabel('y-label')
plt.grid()
plt.legend(loc=2)
plt.show()
```

### Fill Between Plot

```python
x = np.array([1,2,3,4,5,6])
y = np.array([2,4,6,3,1,5])

plt.plot(x,y)
plt.fill_between(x,y, color='g', where=(x>=2) & (x<=5), alpha=.5)
plt.title('Python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
