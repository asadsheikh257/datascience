```python
# ----- Step 2. Load Model and Tokenizer -----
model_name = "bert-base-uncased"
tokenizer = BertTokenizerFast.from_pretrained(model_name)
model = BertForQuestionAnswering.from_pretrained(model_name)
```

```python
# ----- Step 3. Preprocess the Dataset -----
def preprocess_function(examples):
    tokenized_examples = tokenizer(
        examples["question"],
        examples["context"],
        truncation="only_second",  # keep the question intact, truncate the context if needed
        max_length=384,
        stride=128,
        return_overflowing_tokens=True,
        return_offsets_mapping=True,
        padding="max_length"
    )

    sample_mapping = tokenized_examples.pop("overflow_to_sample_mapping")
    offset_mapping = tokenized_examples.pop("offset_mapping")

    tokenized_examples["start_positions"] = []
    tokenized_examples["end_positions"] = []

    for i, offsets in enumerate(offset_mapping):
        sample_index = sample_mapping[i]
        answer = examples["answers"][sample_index]
        answer_text = answer["text"][0]
        start_char = answer["answer_start"][0]
        end_char = start_char + len(answer_text)

        sequence_ids = tokenized_examples.sequence_ids(i)
        # Find start and end token positions for the context (sequence_id==1)
        token_start_index = 0
        while token_start_index < len(sequence_ids) and sequence_ids[token_start_index] != 1:
            token_start_index += 1

        token_end_index = len(sequence_ids) - 1
        while token_end_index >= 0 and sequence_ids[token_end_index] != 1:
            token_end_index -= 1

        # If the answer is not fully inside the span, assign CLS index
        if not (offsets[token_start_index][0] <= start_char and offsets[token_end_index][1] >= end_char):
            cls_index = tokenized_examples["input_ids"][i].index(tokenizer.cls_token_id)
            tokenized_examples["start_positions"].append(cls_index)
            tokenized_examples["end_positions"].append(cls_index)
        else:
            # Otherwise, find the token indices where the answer starts and ends.
            while token_start_index < len(offsets) and offsets[token_start_index][0] <= start_char:
                token_start_index += 1
            tokenized_examples["start_positions"].append(token_start_index - 1)

            while token_end_index >= 0 and offsets[token_end_index][1] >= end_char:
                token_end_index -= 1
            tokenized_examples["end_positions"].append(token_end_index + 1)

    return tokenized_examples

# Apply the preprocessing over the dataset (batched mode)
tokenized_dataset = dataset.map(preprocess_function, batched=True, remove_columns=dataset.column_names)
```

```python
# ----- Step 4. Define Training Arguments and Trainer -----
training_args = TrainingArguments(
    output_dir="./small-bert-qa",
    evaluation_strategy="no",
    learning_rate=3e-5,
    per_device_train_batch_size=1,  # Small batch size for a small dataset
    num_train_epochs=3,             # Few epochs for quick fine-tuning
    weight_decay=0.01,
    logging_steps=5,
    save_steps=5,
    report_to=[],  # Disable wandb logging
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    tokenizer=tokenizer,
)
```

```python
# ----- Step 5. Fine-Tune the Model -----
trainer.train()

# Optionally, save the fine-tuned model
# trainer.save_model("./small-bert-qa")
```
