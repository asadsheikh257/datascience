# **Feature Encoding**


Feature encoding is the process of transforming `categorical features` into `numeric features`. This is necessary because machine learning algorithms can only handle numeric features. There are many different ways to encode categorical features, and each method has its own advantages and disadvantages. In this notebook, we will explore some of the most popular methods for encoding categorical features, such as:

- Label encoding
- Ordinal encoding
- One-hot encoding
- Binary encoding

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
```

```python
le = LabelEncoder()
df['encoded_time'] = le.fit_transform(df['time'])
df.head()
```

```python
df['encoded_time'].value_counts()
```

```python
oe = OrdinalEncoder()
df['encoded_day'] = oe.fit_transform(df[['day']])
df.head()
```

```python
# one hot encoding on day column 
ohe = OneHotEncoder()
ohe.fit_transform(df[['day']]).toarray()
```

```python
# example of one hot encoding
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# example of one hot encoding on titanic data set
onehot_encoder = OneHotEncoder(sparse=False)
embarked_onehot = onehot_encoder.fit_transform(titanic[['embarked']])
embarked_onehot_df = pd.DataFrame(embarked_onehot, columns=onehot_encoder.get_feature_names_out(['embarked']))
titanic = pd.concat([titanic.reset_index(drop=True), embarked_onehot_df.reset_index(drop=True)], axis=1)
titanic.head()
```

### **Answer** :

Feature encoding is a crucial step in preparing data for machine learning models. There are several types of feature encoding techniques:

One-Hot Encoding: This method converts categorical variables into binary vectors, where each category is represented by a binary digit (0 or 1).

Label Encoding: Label encoding assigns a unique integer to each category in a categorical variable. It's simple but might not be suitable for all algorithms because it introduces ordinality.

Ordinal Encoding: Similar to label encoding, ordinal encoding assigns integers to categories, but it considers the order of the categories and assigns integers accordingly.

Binary Encoding: Binary encoding converts categories into binary digits and represents each category with fewer bits compared to one-hot encoding, which can save memory.

Frequency Encoding: This technique replaces categories with the frequency of their occurrence in the dataset.

Target Encoding: Also known as mean encoding, it replaces categories with the mean of the target variable for each category.

Hashing Encoding: Hashing encoding uses a hash function to convert categorical variables into a fixed-length vector, which can be useful when dealing with high cardinality categorical variables.
